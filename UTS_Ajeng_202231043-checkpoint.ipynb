```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk mendeteksi warna merah, biru, dan hijau
def deteksi_warna(gambar):
    # Konversi gambar ke format HSV
    hsv = cv2.cvtColor(gambar, cv2.COLOR_BGR2HSV)

     # Tentukan batas atas dan bawah untuk warna putih
    lower_white = np.array([0, 0, 200], dtype=np.uint8)
    upper_white = np.array([255, 30, 255], dtype=np.uint8)

    # Tentukan rentang warna merah dalam format HSV
    lower_red = np.array([0, 100, 100])
    upper_red = np.array([10, 255, 255])

    # Tentukan rentang warna hijau dalam format HSV
    lower_green = np.array([50, 100, 100])
    upper_green = np.array([70, 255, 255])

    # Tentukan rentang warna biru dalam format HSV
    lower_blue = np.array([110, 50, 50])
    upper_blue = np.array([130, 255, 255])

    # Buat mask untuk setiap warna
    mask = cv2.inRange(hsv, lower_white, upper_white)
    mask_red = cv2.inRange(hsv, lower_red, upper_red)
    mask_green = cv2.inRange(hsv, lower_green, upper_green)
    mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)

    # Gabungkan mask dengan gambar asli
    hasil_red = cv2.bitwise_and(gambar, gambar, mask=mask_red)
    hasil_green = cv2.bitwise_and(gambar, gambar, mask=mask_green)
    hasil_blue = cv2.bitwise_and(gambar, gambar, mask=mask_blue)

    return hasil_red, hasil_green, hasil_blue

# Baca gambar
gambar = cv2.imread('AjengUts.jpg')

# Deteksi warna merah, hijau, dan biru
hasil_red, hasil_green, hasil_blue = deteksi_warna(gambar)

# Tampilkan hasil
plt.figure(figsize=(15, 5))


plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(hasil_red, cv2.COLOR_BGR2RGB))
plt.title('Deteksi Warna Merah')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(hasil_green, cv2.COLOR_BGR2RGB))
plt.title('Deteksi Warna Hijau')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(hasil_blue, cv2.COLOR_BGR2RGB))
plt.title('Deteksi Warna Biru')

plt.show()
```

```python
import cv2
import numpy as np

# Load citra Jupiter
img = cv2.imread('AjengUts.jpg', cv2.IMREAD_COLOR)

# Ubah citra menjadi ruang warna HSV
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Ambil komponen warna H (hue) dari citra HSV
hue_img = hsv_img[:, :, 0]

# Menemukan nilai ambang batas menggunakan metode Otsu
_, threshold = cv2.threshold(hue_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Menampilkan histogram untuk visualisasi
plt.hist(hue_img.ravel(), bins=256, range=[0,256])
plt.title('Histogram Hue (H)')
plt.xlabel('Intensitas')
plt.ylabel('Frekuensi')
plt.show()

# Menampilkan citra dengan penerapan threshold
binary_img = np.zeros_like(hue_img)
binary_img[hue_img > threshold] = 255

cv2.imshow('Binary Image', binary_img)
cv2.waitKey(0)

plt.show()
```

```python
import cv2
import numpy as np

# Load citra Jupiter
jupiter_img = cv2.imread('AjengUts.jpg', cv2.IMREAD_GRAYSCALE)

# Menggunakan Otsu untuk mendapatkan nilai ambang biner
_, binary_thresh = cv2.threshold(jupiter_img, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
_, binary_inv_thresh = cv2.threshold(jupiter_img, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
adaptive_thresh_mean = cv2.adaptiveThreshold(jupiter_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
adaptive_thresh_gaussian = cv2.adaptiveThreshold(jupiter_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

plt.figure(figsize=(12, 6))

plt.subplot(231)
plt.imshow(jupiter_img, cmap='gray')
plt.title('Original Image')

plt.subplot(232)
plt.imshow(binary_thresh, cmap='gray')
plt.title('Binary Threshold')

plt.subplot(233)
plt.imshow(binary_inv_thresh, cmap='gray')
plt.title('Binary Inverted Threshold')

plt.subplot(234)
plt.imshow(adaptive_thresh_mean, cmap='gray')
plt.title('Adaptive Threshold (Mean)')

plt.subplot(235)
plt.imshow(adaptive_thresh_gaussian, cmap='gray')
plt.title('Adaptive Threshold (Gaussian)')

plt.show()
```
